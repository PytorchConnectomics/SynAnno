```python
import sys; sys.path.append("../")
import numpy as np
from matplotlib import pyplot as plt

from synanno.backend.utils import readvol
from synanno.backend.processing import visualize, process_syn
image = readvol("../datasets/image.h5")
label = readvol("../datasets/label.h5")
print(image.shape, image.dtype)
print(label.shape, label.dtype)
```

```python
syn, seg = process_syn(label)
data = visualize(syn, seg, image, sz=128, return_data=True)
```

```python
def show_slices(data, n=8):
    plt.figure(figsize=(16,4))
    n_slices = data.shape[0]
    for i in range(n):
        plt.subplot(1, n, i+1)
        if i >= n_slices:
            plt.imshow(np.zeros_like(data[0]), cmap='gray')
            plt.axis("off")
            continue

        plt.imshow(data[i], cmap='gray')
        plt.axis("off")
        
    plt.tight_layout()
    plt.show()
```

```python
for i, key in enumerate(data.keys()):
    for k in range(2):
        show_slices(data[key][k])
    if i > 3:
        break
```
