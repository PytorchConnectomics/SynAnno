```python
import cloudvolume
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
c3_cloudvolume = cloudvolume.CloudVolume('https://storage.googleapis.com/h01-release/data/20210601/c3', progress=True)
```

```python
# Request the skeleton for a single c3 segment id
skel = c3_cloudvolume.skeleton.get(3896803064)
# Skeleton vertices are in nanometers (not voxels)
skel.vertices

x = skel.vertices[:,0]
y = skel.vertices[:,1]
z = skel.vertices[:,2]

print(x)
```

```python
def plot3D(x, y, z, x_syn, y_syn, z_syn):
    # Create 3D scatter plot

    fig = go.Figure()


    fig.add_trace(go.Scatter3d(
        x=x, 
        y=y, 
        z=z, 
        mode='markers',
        marker=dict(
            size=1,
            color=z,  # Color by z values
            colorscale='Viridis',  # Choose a colorscale
            opacity=0.8
        )
    ))


    fig.add_trace(go.Scatter3d(
            x=x_syn, 
            y=y_syn, 
            z=z_syn, 
            mode='markers',
            marker=dict(
                size=1,
                color='red',  # Color by z values
                opacity=0.8
            )
    ))


    # Update layout
    fig.update_layout(
        title='3D Scatter Plot with Plotly',
        scene=dict(
            xaxis_title='X Axis',
            yaxis_title='Y Axis',
            zaxis_title='Z Axis'
        )
    )

    # Show plot
    fig.show()
```

```python
# load synapse point cloud from csv file
synapses = pd.read_csv('synapse-export_000000000000.csv')
synapses.head()
```

```python
# filter synapses where pre neuron_id is 3896803064
synapses_filtered = synapses[synapses['post_neuron_id'].isin([3896803064])]

x_syn = synapses_filtered['pre_pt_x'].values * 8   
y_syn = synapses_filtered['pre_pt_y'].values * 8
z_syn = synapses_filtered['pre_pt_z'].values * 33
```

```python
plot3D(x, y, z, x_syn, y_syn, z_syn)
```
